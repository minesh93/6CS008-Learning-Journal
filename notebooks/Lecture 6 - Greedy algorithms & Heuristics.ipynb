# 2. Definitions

<b>Heuristic</b> is derived from the Greek word 'heurisko' which means "to find". Its definition in computer science has changed several times in history but currently it is most often used as an adjective to refer to any technique that improves the average-case performance of a problem-solving task, without necessarily finding the best solution. Simply put, it refers to a class of algorithms that provide an estimate of the best solution.

For example, in the 1st example above, a heuristic solution (like the one we'll see below) may tell us that the minimum for function $f(x)$ is reached for $x=0.63$, instead of the optimal $x=0.66$. Similarly, for the 2nd example (route finding) a heuristic solution may give us a route that is shorter than the average, but not the shortest! Heuristics have known significant application in Artificial Intelligence problems.

You may wonder, why would we want to apply heuristic techniques, if they provide sub-optimal solutions. The reason is simple: some problems are too expensive (may take too much time or processing power) to solve exactly. In these cases, heuristics are applied to provide a <i> good enough </i> solution.

<b> Greedy </b> algorithms can be considered as a sub-category of heuristic algorithms. They are based on the idea that while solving a problem we make a decision that seems to be the best decision <i> right now </i> (i.e., locally) without considering if this decision is the best overall (i.e., globally).

For example, in the routing problem imagine we want to travel from Arad to Craiova. There are two ways to do so: 1) via Timisoara or, 2) via Sibiu. A greedy algorithm may choose the first route because: At this particular moment, it looks like the fastest way to move towards our destination is through Timisoara (118km) rather than Sibiu (140km).

Of course, in this case, this is a less than optimal solution since the whole trip in this case will be: $118+111+70+75+120=494km$ instead of $140+80+146=366km$. 

Naturally, this is a trivial example, since we only have 2 options and the calculations are pretty fast. In real-life applications nonetheless you may have 1,000's options and each route calculation may take several minutes/hours. In this case, choosing the greedy solution makes sense, because it is <b> efficient</b> overall.

## 3.1 Finding the minimum of a mathematical function

A first, naive method of finding the minimum of a function is <em>exhaustive search</em>. Simply, we examine the value of function $f(x)$ within a predefined range and keep the smallest one! This approach is guaranteed to find us the global minimum, assuming that it is withint the predefined range we've set.

```python
def exhaustive_search(low,high,step):
    
    if low >= high:
        print 'Wrong parameters'
        return 
    
    min_value = 100000000  #start from very high number
    x_min = 0
    while low < high:
        if f(low) < min_value:
            min_value = f(low)
            x_min = low
        low = low + step
    
    return min_value, x_min
    

%time min, x = exhaustive_search(-100,100,0.00001)
print 'Minimum reached at: %.3f. It is: %.3f' % (x,min)
    
```

## 3.2 Gradient Descent

<b> Gradient Descent</b> is an example of hill-climbing algorithm for finding the minimum of mathematical functions. It is based on a very simple idea:
<ol>
    <li> Initially, we choose a random point in the plot of the function we want to minimize
    <li> Next we examine the plot of the function at the point we are at and see  at which direction the function takes a lower value
    <ul>
        <li> Actually, in multidimensional settings, we find the direction with the steepest descent
        </ul>
     <li> We move a bit in that direction and repeat the whole process from 1 until we've reached a 'valley' of the function plot.

</ol>

Examine the figures from example 1 to make sure you understand the high-level process.

Gradient Descent is an example of a greedy algorithm. It moves towards the closest valley in the plot of the function it tries to minimize, without being aware if there are other valleys that go deeper than the one it is currently going towards. Subsequently, it is able to find local minima of functions, but not necessarily global ones. 

<em> Note:</em> if the function is convex, gradient descent is actually guaranteed to find the global minimum. This is beyond the scope of this module, but it's good to know.

### 3.1.1 Technicalities

Technically, gradient descent can be summarised as follows, assuming we want to minimize function $f(x)$:

Repeat until convergence {

$x := x - \alpha \frac{\Delta f}{\Delta x}$

}

where $\frac{\Delta f}{\Delta x}$ is the partial derivative of function $f(x)$ to $x$ (don't worry we'll give that to you). Note that the symbol $:=$ simply means replace (the equivalent of $=$ in programming, not in mathematics). Let's see how this works for function $f(x)=3x^2-4x+4$:

```python
def f_derivative(x):
    return 6*x-4

def gradient_descent(alpha,precision):
      
    x_values = []                                  # we'll use this to store all the x_values examined
    x_old = 100                                    # start from random value, 
    x_new =  x_old - alpha * f_derivative(x_old)   # do the first step, before going to the while loop
    
    x_values.append(x_old)
    x_values.append(x_new)
    print 'First step: %.3f' % x_new
    while abs(x_new - x_old) > precision:
        x_old = x_new
        x_new = x_old - alpha * f_derivative(x_old)
        x_values.append(x_new)

    
    return x_new, x_values

%time minimum, examined_values = gradient_descent(0.001,0.001) 
print 'Minimum reached at: %.3f. It is: %.3f'  %  (minimum,f(minimum))
```

In the plot below, we show the decreasing values of $f(x)$ for different values of x during gradient descent.

```python
f_values = []
for index,x in enumerate(examined_values):
    f_values.append(f(x))
    
plt.plot(range(len(f_values)),f_values)                
plt.ylabel('f(x)')              # adding visual elements
plt.xlabel('Iteration')
plt.title('Gradient descent')
plt.ylim(0,f_values[0])
plt.show
```

Lastly, we investigate the issue of route finding. We'll discuss a simple approach here; given a weighted graph (like the map above), go from one location to another by simply following the smaller distance to the next connected node (i.e., city).

```python
world_map = {'Orades': {'Zerind':71, 'Sibiu':151},
          'Zerind': {'Arad':75, 'Orades':71},
          'Arad': {'Timisoara':118, 'Zerind':75, 'Sibiu':140},
          'Timisoara': {'Lugoj': 111, 'Arad':118},
          'Lugoj': {'Mehadia':70, 'Timisoara':111},
          'Mehadia':{'Dobreta':75, 'Lugoj':70},
          'Dobreta': {'Craiova':120,'Mehadia':70},
          'Craiova': {'Rimnicu Vilcea':146, 'Pitesti':138},
          'Rimnicu Vilcea': {'Sibiu':80, 'Pitesti':97, 'Craiova':138},
          'Sibiu': {'Arad':140, 'Oradea':151, 'Fagares':99, 'Rimnicu Vilcea':80},
          'Fagares': {'Bucharest':211, 'Sibiu':99},
          'Pitesti': {'Rimnicu Vilcea':97, 'Bucharest':101, 'Craiova':138},
          'Bucharest': {'Giurgiu':90, 'Urziceni':85, 'Fagares':211, 'Pitesti':101},
          'Urziceni': {'Hirsova':98, 'Vaslui':142, 'Bucharest':85},
          'Giurgiu': {'Bucharest':90},
          'Hirsova': {'Eforie':86, 'Urziceni':98},
          'Eforie': {'Hirsova':86},
          'Vaslui': {'Iasi':92, 'Urziceni':142},
          'Iasi' : {'Neamt':87, 'Vaslui':92},
          'Neamt': {'Iasi':87},
          }
```

```python
from sets import Set

def greedy_route(current_map,start, end, route = [], visited = Set(), total_distance = 0):
    
    visited.add(start)   # visited will contain all the cities we've already visited, so that we avoid going in circles!
    
    if start == end:     # in case our start and end are the same city
        print 'Start city is the same as end!'
        return route,0
    
    if not current_map.has_key(start) or not current_map.has_key(end):   # in case the start or the end aren't in our current_map
        print 'No possible route'
        return
    
    else:                                          # if all's well, let's find the closest city to our current 'start'
        min_distance = 100000                      # temporarily set the minimum distance from 'start' city to high number                              
        closest_city = '';                                                               
        
        for dest in current_map[start].keys():     #iterate through the connected cities and get/print their distance                  
            print '%s->%s: %d' % (start, dest, current_map[start][dest])    
            
            if dest == end:                       # if we've found our final destination, even if it's not the closest city
                route.append(dest)
                total_distance = total_distance + current_map[start][dest]
                return route, total_distance
            
            elif current_map[start][dest] < min_distance and dest not in visited:   # if the city we are examining now is closer than the current minimum
                min_distance = current_map[start][dest]                             # AND also we haven't visited the city already
                closest_city = dest
        
        if closest_city == '':                            # if we reach a dead-end
            print 'Dead end!'
            return [],0
        
        route.append(closest_city)                         # add the closest_city to the route and the set of visited cities
        visited.add(closest_city)                
        total_distance = total_distance + min_distance      # also add its distance to total distance
        print 'Closest  (& not visited): %s' % closest_city
        print 'Distance so far: %d \n' % total_distance
        
        if closest_city == end:
            return route,total_distance
        else:
            route, total_distance = greedy_route(current_map,closest_city,end,route,visited,total_distance)
          
        return route,total_distance
            
        
itenary, distance = greedy_route(world_map, 'Arad','Bucharest')
print 'Greedy route is:', itenary
print 'Total km:', distance
```
